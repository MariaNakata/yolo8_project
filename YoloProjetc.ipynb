```python
# Instalar a biblioteca ultralytics
!pip install ultralytics

#  Montar o Google Drive
from google.colab import drive
drive.mount('/content/drive')

# O código principal do YOLOv8
import os
from ultralytics import YOLO
from IPython.display import Image, display

def detect_objects_in_images(image_paths, model_name='yolov8n.pt'):
    """
    Detects and classifies objects in a list of images using YOLOv8.
    """
    try:
        model = YOLO(model_name)
        print(f"Modelo {model_name} carregado com sucesso.")

    except Exception as e:
        print(f"Erro ao carregar o modelo YOLOv8: {e}")
        print("Por favor, verifique se 'ultralytics' está instalado e se o nome do modelo está correto.")
        return

    output_dir = "/content/drive/MyDrive/yolo_project/deteccoes"
    os.makedirs(output_dir, exist_ok=True)
    print(f"As imagens processadas serão salvas em: {output_dir}")
    print("\n--- Imagens processadas e exibidas ---")

    for image_path in image_paths:
        if not os.path.exists(image_path):
            print(f"Aviso: Imagem não encontrada em '{image_path}'. Pulando...")
            continue

        print(f"\nProcessando imagem: {image_path}")
        try:
            results = model(image_path)

            for result in results:
                boxes = result.boxes

                if len(boxes) == 0:
                    print("Nenhum objeto detectado nesta imagem.")
                else:
                    for box in boxes:
                        xyxy = box.xyxy[0].tolist()
                        x1, y1, x2, y2 = [round(coord) for coord in xyxy]
                        confidence = box.conf[0].item()
                        class_id = box.cls[0].item()
                        class_name = model.names[int(class_id)]

                        print(f"  Objeto Detectado: {class_name}")
                        print(f"    Coordenadas (x1, y1, x2, y2): ({x1}, {y1}, {x2}, {y2})")
                        print(f"    Confiança: {confidence:.2f}")

                output_filename = os.path.join(output_dir, f"detected_{os.path.basename(image_path)}")
                result.save(filename=output_filename)
                print(f"Imagem com detecções salva em: {output_filename}")

                try:
                    display(Image(filename=output_filename, width=600)) # <--- Exibe a imagem
                except Exception as display_e:
                    print(f"Erro ao exibir a imagem no Colab: {display_e}")


        except Exception as e:
            print(f"Erro ao processar a imagem '{image_path}': {e}")
            print("-" * 30)


if __name__ == "__main__":
    input_images_folder = "/content/drive/MyDrive/yolo_project"

    image_files = []
    if os.path.exists(input_images_folder):
        if os.path.isdir(input_images_folder):
            image_files = [
                os.path.join(input_images_folder, f)
                for f in os.listdir(input_images_folder)
                if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))
            ]
            if not image_files:
                print(f"Nenhuma imagem encontrada na pasta '{input_images_folder}'. Por favor, adicione fotos lá.")
        else:
            print(f"Erro: '{input_images_folder}' não é um diretório. Certifique-se de que o caminho aponta para uma PASTA, não para um arquivo.")
            print(f"Caminho atual: {input_images_folder}")
            image_files = []
    else:
        print(f"A pasta '{input_images_folder}' não foi encontrada no seu Google Drive.")
        print("Por favor, crie-a e adicione suas imagens.")
        print("Caminho esperado: /content/drive/MyDrive/yolo_project")
        image_files = []

    chosen_model = 'yolov8n.pt'

    if image_files:
        detect_objects_in_images(image_files, chosen_model)
    else:
        print("Nenhuma imagem para processar. Verifique o caminho e o conteúdo da pasta.")
```
